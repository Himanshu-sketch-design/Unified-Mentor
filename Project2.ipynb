Task 2 :

1) Employee Age and Salary

```python
# Create a Scatter plot of the 'Age' and 'Employee Salary' features

plt.figure(figsize = (10, 5))

sns.scatterplot(data = data, x = 'age', y = 'salary', color = 'blue', edgecolor = 'linen', alpha = 0.5)

plt.title('Scatter plot of the Age v/s Employee Salary')
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')

plt.show()
```

Observations : 

There is a general relationship between Employee Age and Salary

As age increases salary also increases, though there are some outliers too

2) Healthy Eating and Active Lifestyle

```python
# Create a Scatter plot of the 'Healthy Eating' and 'Active Lifestyle' features

plt.figure(figsize = (10, 5))

sns.scatterplot(data = data, x = 'healthy_eating', y = 'active_lifestyle', color = 'green', edgecolor = 'linen', alpha = 0.5)

plt.title('Scatter plot of the Healthy Eating v/s Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')

plt.show()
```

Observations : 

As Healthy Eating increases, the Active Lifestyle also increases 

3) Healthy Eating and Salary

```python
# Create a Scatter plot of the 'Healthy Eating' and 'Salary' features

plt.figure(figsize = (10, 5))

sns.scatterplot(data = data, x = 'healthy_eating', y = 'salary', color = 'green', edgecolor = 'linen', alpha = 0.5)

plt.title('Scatter plot of the Healthy Eating v/s Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')

plt.show()
```

Observations :
 
As Healthy Eating increases, Salary also increases

4) Groups

```python
# Create a Count plot of the 'groups' features

plt.figure(figsize = (10, 5))

sns.countplot(data = data, x = 'groups', color = 'red', edgecolor = 'linen', alpha = 0.5)

plt.title('Count Plot of Groups')
plt.xlabel('Groups')
plt.ylabel('Count')

plt.show()
```

Observations :

Most of the employee belongs to group 'A' and group 'O' with group 'A' having maximum frequency

5) Salary

```python
# Create a histogram of the 'Salary' features

plt.figure(figsize = (11, 6))

sns.histplot(data = data, x = 'salary', color = 'orange', edgecolor = 'linen', alpha = 0.7, bins = 10)

plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Count')

plt.show()
```

Observations : 

As we can see that salaries are not uniform, which is not necessarily a discrepancy

We observe that there are employees with salaries on either extremites of the histogram

Based on the histogram, we can see that the count of his=ghest earning employees is very less

Task 3 :

```python
# An employee of healthy eating greater than 8 

sub1 = data[data['healthy_eating'] > 8]
```

```python
# Salary less than 8

sub2 = data[data['salary'] < 1000]
```

```python
# Employee with healthy eating greater than 8 an with salary less than 1000

sub = data[(data['healthy_eating'] > 8) & (data['salary'] < 1000)]
```

```python
# Output : 

print("1st Output : \n\n",sub1, "\n")
print("2nd Output : \n\n",sub2, "\n")
print("3rd Output : \n\n",sub, "\n")
```

Observations from Task 3 :

The only employee seemingly facing a discrepancy in salary as compared to healthy eating is employee whose id = 8

The employee with id having 26 has a salary of 700
