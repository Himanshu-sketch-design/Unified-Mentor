1) Fixed Acidity : 

```python
# Create a histogram of the 'fixed acidity' features

plt.figure(figsize= (11, 6))

sns.histplot(data = data, x = 'fixed acidity', color = 'blue', edgecolor = 'purple', alpha = 0.5, bins = 5)

plt.title('Histogram of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')

plt.show()
```

Observations : 

We see the histogram is normally distributed.

The maximum count for fixed acidity lies in between 6 to 8.

Calculating the Mean and Median of the data

```python
# Mean : 

round(data['fixed acidity'].mean(), 2)
```

```python
# Median : 

round(data['fixed acidity'].median(), 2)
```

```python
# Applying the Mean and Median by creating a histogram of the 'fixed acidity' features

plt.figure(figsize= (11, 6))

sns.histplot(data = data, x = 'fixed acidity', color = 'orange', edgecolor = 'green', alpha = 0.5, bins = 5)

plt.title('Histogram of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')

plt.vlines(data['fixed acidity'].mean(), ymin = 0, ymax = 4000, colors = 'blue', label = 'Mean')
plt.vlines(data['fixed acidity'].median(), ymin = 0, ymax = 4000, colors = 'red', label = 'Median')

plt.legend()

plt.show()
```

Observations : 

Mean and Median are very close to each other. So, we can choose any one of them as the measure of central tendency

2) Volatile Acidity

```python
# Create a histogram of the 'volatile acidity' features

plt.figure(figsize= (11, 6))

sns.histplot(data = data, x = 'volatile acidity', color = 'red', edgecolor = 'purple', alpha = 0.5, bins = 5)

plt.title('Histogram of Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')

plt.show()
```

Observations :

It's shown from above the histogram that the chart is not well distributed, it's skewed a little towards the right

Theredfore, we check distribution using distplot function

Skewness : measure of the asymetric distribution

It can be determined by : 

        > if skewness(S) = 0, then distribution is normally distributed
        > if skewness(S) > 0, then distribution is possitively skewed
        > if skewness(S) < 0, then distribution is negatively skewed

```python
# Plot distplot using 'volatile acidity' features 

plt.figure(figsize = (11, 6))

sns.distplot(data['volatile acidity'], color = 'blue')

plt.title('Distplot of Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')

plt.show()
```

Observation :

The above plot shows normal distribution

The normal distribution is described by the mean and the standard deviation

The normal distribution is often referred to as a 'bely curve' because of it's shape : 

        > The median and the mean are equal
        > It has only one node
        > It's symmetric, meaning it decreases the same amount on the left and the right of the centre

```python
# Calculate Skewness of 'volatile acidity'

data['volatile acidity'].skew()
```

```python
# Showing the Mean and Median by creating a histogram of the 'volatile acidity' features

plt.figure(figsize= (11, 6))

sns.histplot(data = data, x = 'volatile acidity', color = 'green', edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title('Histogram of Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')

plt.vlines(data['volatile acidity'].mean(), ymin = 0, ymax = 4000, colors = 'blue', label = 'Mean')
plt.vlines(data['volatile acidity'].median(), ymin = 0, ymax = 4000, colors = 'red', label = 'Median')

plt.legend()

plt.show()
```

Observation : 

The mean and the median are close to each other and the difference between them are very small.

So we can safely choose mean as the measure of central tendency

Task 3 :

```python
# Create a new Pandas Series that contains the details of the representative quality for the different types of acid

acid = pd.Series(index = ['Fixed Acidity', 'Volatile Acidity', 'Citric Acid', 'Quality'], 
                 data = [data['fixed acidity'].mean(), data['volatile acidity'].mean(), data['citric acid'].mean(), data['quality'].value_counts().index[0]])
```

```python
# print 

round(acid, 2)
```

Observation from Task 3 : 

The representative acid for the quality are as follows : 

    1) The mean value of the fixed acidity would be 6.854
    2) The mean value of the volatile acidity would be 0.2782
    3) The mean value of the citric acid would be 0.3341
    4) The quality would be 6.854
